## âœ… Supervised Learning

Supervised learning involves training algorithms on a dataset that includes both **input features (X)** and **target outputs (Y)**. The algorithm learns a mapping from inputs to outputs using labeled examples.

| Input (X)             | Output (Y)                 | Application               |
|-----------------------|----------------------------|---------------------------|
| Email                 | Spam (0) / Not Spam (1)     | Spam Filtering            |
| Audio                 | Text transcripts            | Speech Recognition        |
| English Text          | Spanish Text                | Machine Translation       |
| Ad, User Info         | Relevant Ad Clicks          | Online Advertising        |
| Image, Radar Info     | Object/Cars Positions       | Self-driving Cars         |
| Product Image         | Defect (Yes/No)             | Visual Inspection         |

### ðŸ” Types of Supervised Learning

- **Regression Algorithms**:  
  Predict continuous values (e.g., house prices, temperatures).  
  > Example: Predicting tomorrow's temperature.

- **Classification Algorithms**:  
  Predict discrete categories or labels.  
  > Example: Determining if a tumor is malignant or benign.

---

## ðŸ”“ Unsupervised Learning

Unsupervised learning refers to algorithms that analyze and learn from data **without labeled outputs**. The algorithm tries to find hidden patterns or intrinsic structures in the input data.

### ðŸ§© Types of Unsupervised Learning

- **Clustering**:  
  Groups similar data points based on their characteristics.  
  > Example: Grouping people based on daily physical activity patterns into categories like "active" and "sedentary" (without predefined labels).

- **Anomaly Detection**:  
  Identifies rare or unusual data points.  
  > Example: Detecting fraudulent transactions in financial systems.

- **Dimensionality Reduction**:  
  Compresses large datasets into smaller representations using fewer variables, while retaining as much information as possible.  
  > Example: Reducing a dataset of 100 features into 10 principal components.
  

# ðŸ“Š Linear Regression

Linear Regression is one of the simplest and most widely used algorithms in machine learning. It is used to model the relationship between a **dependent variable (target)** and one or more **independent variables (features)** by fitting a linear equation to the observed data.

---

## ðŸ”¢ 1. What is Linear Regression?

> Linear Regression tries to find the best-fitting straight line through the data points.

For one feature (simple linear regression), the hypothesis function is:

$$
f_{w,b}(x) = wx + b
$$

Where:
- \( f_{w,b}(x) \): Predicted output
- \( w \): Weight (slope)
- \( b \): Bias (y-intercept)
- \( x \): Input feature

If we ignore bias (set \( b = 0 \)), the simplified hypothesis becomes:

$$
f_w(x) = wx
$$


## ðŸ“Š Training Set and Hypothesis Relation

A **training set** is a collection of labeled examples:

| $x^{(i)}$ (input) | $y^{(i)}$ (output) |
|------------------|--------------------|
| 2                | 50                 |
| 4                | 70                 |
| 6                | 90                 |

For each training example $(x^{(i)}, y^{(i)})$, the hypothesis function computes:

$$
\hat{y}^{(i)} = f_{w,b}(x^{(i)}) = wx^{(i)} + b
$$

The goal of training is to minimize the **difference** between the actual label $y^{(i)}$ and the predicted value $\hat{y}^{(i)}$ across all training examples.

---

```python
import numpy as np
import matplotlib.pyplot as plt

# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of training examples
print(f"x_train.shape: {x_train.shape}")
# One can also use the Python len() function as well 
# m = len(x_train)
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
i = 0 # Change this to 1 to see (x^1, y^1)

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

## ðŸ“Œ 2. Cost Function

The cost function helps measure how well the linear regression model fits the data.



$$
J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} \left(f_{w,b}(x^{(i)}) - y^{(i)}\right)^2
$$

Simplified when bias \( b \) is 0:

$$
J(w) = \frac{1}{2m} \sum_{i=1}^{m} \left(wx^{(i)} - y^{(i)}\right)^2
$$

Goal: Minimize \( J(w, b) \)

## âš™ï¸ 5. Gradient Descent

Gradient Descent is an optimization algorithm that updates weights to minimize the cost:

$$
w := w - \alpha \frac{\partial}{\partial w} J(w, b) \\
$$
$$
b := b - \alpha \frac{\partial}{\partial b} J(w, b)
$$

## ðŸ§® 9. Evaluation Metrics

- **Mean Squared Error (MSE)**:  
  $$
  MSE = \frac{1}{m} \sum_{i=1}^{m} (y^{(i)} - \hat{y}^{(i)})^2
  $$

- **Root Mean Squared Error (RMSE)**:  
  $$
  RMSE = \sqrt{MSE}
  $$

- **Mean Absolute Error (MAE)**:  
  $$
  MAE = \frac{1}{m} \sum_{i=1}^{m} \left| y^{(i)} - \hat{y}^{(i)} \right|
  $$

- **RÂ² Score**:  
  $$
  R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
  $$
