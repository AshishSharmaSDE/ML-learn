```python
# Importing pandas library for data analysis
import pandas as pd

print('Import Setup Done')
```

```python
# Reading CSV file into a DataFrame
# import data from csv files into dataframes and remove the index columns that is auto generated with values on 0-n in rows
# downloaded data from stack overflow survey page: https://survey.stackoverflow.co/
# For example, you can see in this dataset that the CSV file has a built-in index, which pandas did not pick up on automatically. 
# To make pandas use that column for the index (instead of creating a new one from scratch), we can specify an index_col
surveys = pd.read_csv('data/survey_results_public.csv', index_col=0)
print('Loading data!!')
```

```python
# Displaying the first 5 rows of the DataFrame
# Display first 5 records in dataframe using head() Pass the numbers of records you want to fetch like head(15)
surveys.head()
# Display last 5 records in dataframe using tail() Pass the numbers of records you want to fetch like tail(15)
surveys.tail()
```

```python
# Checking the number of rows and columns in the DataFrame
# too see the no. of rows and column you can use Shape
surveys.shape
```

```python
# Code cell
# to select any column data we cam directly use surveys.CodingActivities OR surveys['CodingActivities']
# extracting a column from data frame returns a series only
# Datafame is like a table whereas the series is likle an array or list.
activities = surveys.CodingActivities
print(type(activities))
```

```python
# Accessing specific rows or columns using .loc/.iloc
# now in order to access spectific records we use
# df.iloc[0]  only works on the index of the columns not on the names
# df.loc[:,0]
# pandas also uses slicing using : for example select 0-10 records
surveys.iloc[0:10]
# it can use a list or numbers as well which will return thw rows numbered in the list
surveys.iloc[[1,3,5,7,9], 3] # ther is an additional param which can be passed as index of column either 0,1,23
```

```python
# Accessing specific rows or columns using .loc/.iloc
# in order to access the records based on the names of cilumns we use
surveys.loc[:10, 'RemoteWork']
# here the first param will the number of records you wanto to access 2nd param will be the name os columns you wnat it could be a list as well
cols = ['CodingActivities', 'EdLevel','LearnCode']
surveys.loc[:10,cols]# get all the records with given cols only again slicing can be done here like we slice in string where right part of slice is not inclued
```

```python
# Accessing specific rows or columns using .loc/.iloc
# conditional selection 
# we can utilise == (Equal operator) & (AND operator) | (OR operator)
# other operator as well >,<,>=,<=
surveys.loc[surveys.CodingActivities == 'Hobby']
surveys.loc[(surveys.CodingActivities == 'Hobby') & (surveys.RemoteWork == 'Remote') | (surveys.JobSat > 5 )]

# we have other methods as well like
surveys.loc[surveys.CodingActivities.isin(['Hobby', 'I don’t code outside of work'])]

surveys.CodingActivities.notnull()

# we can assign values to existing df like
surveys['CodingActivities'] = 'xyz' # it will alter all the vvalues of CodingActivities to xyz.
```
